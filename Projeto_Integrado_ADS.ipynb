```python
#Dicionário para tipos de movimentações
tipos_movimentacoes = {
    'entrada': 1,
    'saida': 2
}

#Lista para armazenar os produtos
produtos = []
categorias = []
movimentacoes = []

#criar classes
class Produto:
    def __init__(self,id,nome,categoria,quantidade):
        self.id = id
        self.nome = nome
        self.quantidade = quantidade
        self.categoria = categoria

class Categoria:
  def __init__(self,id,nome):
    self.id = id
    self.nome = nome

class Movimentacao:
  def __init__(self,id,produto,tipo,quantidade,data):
    self.id = id
    self.produto = produto
    self.tipo = tipo
    self.quantidade = quantidade
    self.data = data

#função para cadastrar produto
def cadastrar_produto(produtos):
  while True:
    try:
      id = int(input('Digite o id do produto:'))
      nome = input('Digite o nome do produto:')
      categoria = input('Digite a categoria do produto:')
      quantidade = int(input('Digite a quantidade do produto:'))
      if quantidade < 0:
        print('Quantidade inválida.')
        continue
      break
    except ValueError:
      print('Entrada inválida.')

  novo_produto = Produto(id,nome,categoria,quantidade)
  produtos.append(novo_produto)

#função para consultar produto
def consultar_produto(produtos,id):
  if not produtos:
    return "Nenhum produto cadastrado."
  for produto in produtos:
    if produto.id == id:
      return produto
  return "Produto não encontrado."


#função para cadastrar categoria
def cadastrar_categoria(categorias,id,nome):
  nova_categoria = Categoria(id,nome)
  categorias.append(nova_categoria)

#função para consultar categoria
def consultar_categoria(categorias,id):
  for cate in categorias:
    if cate.id == id:
      return cate
  return "Categoria não encontrada."

#registrar entrada de produto
def entrada_produto(movimentacoes,id,produto,quantidade,data):
  nova_movimentacao = Movimentacao(id,produto,tipos_movimentacoes['entrada'],quantidade,data)
  movimentacoes.append(nova_movimentacao)
  produto.quantidade += quantidade

#registrar saída de produto
def saida_produto(movimentacoes,id,produto,quantidade,data):
  nova_movimentacao = Movimentacao(id,produto,tipos_movimentacoes['saida'],quantidade,data)
  movimentacoes.append(nova_movimentacao)
  produto.quantidade -= quantidade

#gerar relatório de estoque
def relatorio_estoque(produtos):
  relatorio = []
  for produto in produtos:
    relatorio.append({
        'id': produto.id,
        'nome': produto.nome,
        'quantidade': produto.quantidade
    })
  return relatorio

#consultar movimentações
def consultar_movimentacao(movimentacoes):
  historico = []
  for movi in movimentacoes:
    historico.append({
        'id': movi.id,
        'produto': movi.produto.nome,
        'tipo': movi.tipo,
        'quantidade': movi.quantidade,
        'data': movi.data
    })
  return historico
```
